### Task: Take any data set and plot different kind of graphs

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/LavanyaPolamarasetty/Datasets/master/Classification/Social_Network_Ads.csv")
df.head()
```

```python
# Draw Line Plot of Age and Estimated Salary
import matplotlib.pyplot as plt
plt.plot(df["Age"][:10])
plt.title("Line Plot")
plt.xlabel("Age")
plt.ylabel("Number of people")
plt.show()
```

```python
# Draw Scatter Plot
plt.scatter(df["Age"],df["EstimatedSalary"])
plt.title("Scatter Plot")
plt.xlabel("----> Age of the Employee")
plt.ylabel("----> Estimated Salary of the Employee")
plt.show()
```

```python
#Histogram
plt.hist(df["Age"], width = 0.6 ,color = "r")
plt.title("Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency count of the Age")
plt.show()
```

```python
df["Purchased"].value_counts()
```

```python
#bar graph
plt.bar(df["Gender"],df["EstimatedSalary"])
plt.title("Bar graph")
plt.xlabel("Gender")
plt.ylabel("EstimatedSalary")
plt.show()
```

```python
df["Gender"].value_counts()
```

```python
# Pie Chart 
plt.pie(df["EstimatedSalary"][:10],labels = df["Gender"][:10],radius = 1.5, autopct = "% 1.1f%%")
plt.show()
```

```python
# Box Plot
plt.boxplot(df["EstimatedSalary"])
plt.title("Box Plot")
plt.xlabel("Estimated Salary")
plt.show()
```
